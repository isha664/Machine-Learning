# import some  important libraries 


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.preprocessing import LabelEncoder
```

## Load the dataset

```python
df=pd.read_csv("C:\\Users\\Isha\\Desktop\\AI class\\DS dataset\\50_Startups  multiple linear regressiion.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
X = df[['R&D Spend', 'Administration', 'Marketing Spend', 'State']]
y = df['Profit']
```

### Use encoding method 

```python

label_encoder = LabelEncoder()

st = label_encoder.fit_transform(df['State'])
```

### Split the model into train test  

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

### Initialize the Model

```python
regression=LinearRegression()
regression
```

### Fit the model

```python
model=regression.fit(X_train,y_train)
model
```

### Predict on the test set

```python
y_pred=model.predict(X_test)
y_pred
```

### Evaluate the value 

```python

squared= mean_squared_error(y_pred,y_test)
squared
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color='black', alpha=0.7)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='green')
plt.xlabel('Actual Profit')
plt.ylabel('Predicted Profit')
plt.title('Predicted vs Actual Profit')
plt.show()
```
